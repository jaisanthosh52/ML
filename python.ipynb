```python
import matplotlib.pyplot as plt

# Calculate percentages
percentages = round(data['total_bill'].value_counts(normalize=True) * 100, 2)
percentages.bar_label(percentages.containers[0])

# Create figure
fig, ax = plt.subplots(figsize=(10,15))

# Plot bar chart
bars = ax.bar(percentages.index, percentages.values,
              color=['green', 'red', 'blue'])

# Add labels on top of bars
ax.bar_label(bars, fmt='%.1f%%')

# Set x-ticks
plt.xticks(rotation=45)

# Add grid only on Y-axis
ax.grid(axis='y', color='black', ls=':', dashes=(1,5))

# Labels and title
ax.set_xlabel("Total Bill")
ax.set_ylabel("Percentage (%)")
ax.set_title("Percentage Distribution of Total Bill")

plt.show()
```
